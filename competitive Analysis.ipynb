```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn import preprocessing
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
data=data.replace(['a','b','c','d'],[1,2,3,4])
data
```

```python
data['StateHoliday'] = data['StateHoliday'].astype(int)
```

```python
data['Date'] = data['Date'].astype('datetime64[ns]').astype("int64")
```

```python
data=data.drop('PromoInterval',axis=1)
data
```

```python
x=data.drop(['Sales'],axis=1)
y=data['Sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
scaler = preprocessing.StandardScaler()
```

```python
x_train_scalled = scaler.fit_transform(x_train)
```

```python
yp_train=model.predict(x_train_scalled)
yp_train.shape
```

```python
model.score(x_train,y_train)
```

```python
mae=metrics.mean_absolute_error(y_train,yp_train)
mae
```

```python
mse = metrics.mean_squared_error(y_train, yp_train)
mse
```

```python
rmse = np.sqrt(metrics.mean_absolute_error(y_train, yp_train))
rmse
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score=r2_score(y_train,yp_train)
r2_score
```

```python
DataFrame=pd.DataFrame({'MAE': [mae], 'MSE': [mse], 'RMSE': [rmse],'r2_score':[r2_score]})
DataFrame
```
